```python
import numpy as np
import csv
import sys
sys.path.insert(1, '../../Scripts')
from measfilter import getData0
from gatesimu import *
```

```python
file_address = 'T1/'
m = 200
with open(file_address + 'Readout_{}{}Q{}.csv'.format(m, 'X', 1), mode='r') as measfile:
    reader = csv.reader(measfile)    
    data1 = np.asarray([row for row in reader][0])
with open(file_address + 'Readout_{}{}Q{}.csv'.format(m, 'X', 2), mode='r') as measfile:
    reader = csv.reader(measfile) 
    data2 = np.asarray([row for row in reader][0])
```

```python
gm = GMFilter([1,2], m,'X', 'Noise/', file_address)
#gm.post_from_file()
gm.inference()
```

```python
p0s1 = getData0(data1, 128, 1)
p0s2 = getData0(data2, 128, 2)
Q1BJW = gm.filter_mean(p0s1, 1)
Q1SB = gate_denoise(m, p0s1, [0.9221484,  0.89385907, 0.00936743])

Q2BJW = gm.filter_mean(p0s2, 2)
Q2SB = gate_denoise(m, p0s2, [0.9213698,  0.88708473, 0.00596481])

import matplotlib.pyplot as plt
plt.hist([Q1BJW, Q1SB], label = ['BJW', 'SB'])
plt.xlabel('Probability of measuring 0')
plt.ylabel('Counts')
plt.legend()
plt.savefig('denoised-Qubit1.jpg')
plt.show()
plt.hist([Q2BJW, Q2SB], label = ['BJW', 'SB'])
plt.xlabel('Probability of measuring 0')
plt.ylabel('Counts')
plt.legend()
plt.savefig('denoised-Qubit2.jpg')
plt.show()
```
