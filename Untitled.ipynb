```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
q1 = pd.read_csv("Paper Experiment - U3 Gates/T1/Qubit1.csv")
q2 = pd.read_csv("Paper Experiment - U3 Gates/T1/Qubit2.csv")
```

```python
sns.kdeplot(data=q1, x="y")
plt.show()
```

```python
sns.kdeplot(data=q2, x="y",label="Q2 Data")
sns.kdeplot(data=q1, x="y",label="Q1 Data")
plt.legend()
plt.xlabel("Pr(Meas. 0)")
plt.show()
```
